```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('seaborn-whitegrid')


datos = sns.load_dataset('planets')
datos.head()
```

```python
print('El dataset tiene', datos.shape[0],'filas y', datos.shape[1],'columnas')
```

```python
print('las columnas del dataset son:', datos.columns)
```

```python
for columna in datos:
    print(columna, datos[columna].unique())
```

```python
print('La columna method corresponde a la manera o técnica con la que se descubrió el exoplaneta, estas son:', datos['method'].unique())
print()
print('La columna year corresponde al año en que se descubrio el exoplaneta, los años con descubrimientos son:', datos.sort_values('year')['year'].unique())
```

```python
sns.boxplot(x="year", y="method", data=datos,
            whis=[0, 100], width=0.4, palette="vlag")
```

```python
pltdata = datos.year.value_counts().reset_index()

fig, ax = plt.subplots(figsize=(8, 5))
plt.plot(pltdata.sort_values('index')['index'], pltdata.sort_values('index')['year'], '-', color='black');
plt.plot(2011,185, marker='o', color= 'red')
plt.plot(1989,1, marker='o', color= 'blue')
style = dict(size=10, color='black')

ax.text(2011, 185, "Max: 185 planetas (2011)   ",ha='right', **style)
ax.annotate("Primer planeta reportado en la lista (1989)", xy=(1989, 1),  xycoords='data', bbox=dict(boxstyle="round", fc="none", ec="gray"),
            xytext=(20, 60), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=0.2"))
ax.set(title='Cantidad de exoplanetas por año 1989-2014',
       ylabel='Cantidad de planetas')
```

```python
datos.corr()
```

```python
sns.heatmap(datos.corr(), cmap = 'brg')
```

De acuerdo a la gráfica y la tabla anterior, la correlación más alta es entre la masa y la distancia entre los planetas (color rojo vivo), siendo esta 0.274082. Aun así, siendo la máxima, es una correlación muy baja, es decir, en general los datos están poco relacionados entre si.

```python
sns.pairplot(datos, hue="method")
```

De acuerdo a la matriz anterior, al fijarnos en la gráfica que relaciona el año (eje x) y la distancia (eje y), podemos observar por ejemplo que conforme pasan los años hay menos puntitos azules y hay más puntitos de otros colores, es decir, los métodos nuevos para descubrir planetas, desplazan en gran parte a los métodos viejos. También, podemos ver que conforme pasan los años, las distancias aumentan, es decir, cada vez los instrumentos para detectar planetas pueden observan más lejos. Combinando ambas afirmaciones: conforme pasan los años los nuevos métodos para descubrir planetas permiten captar planetas más lejanos, por lo que los métodos viejos quedan parcialmente obsoletos.
